# Profiles of temperature and salinity for both observation and model

In this notebook I will draw the profiles against the depth of temperature and salinity 
  - first for a single observation and one run
  - second for a single observation and the ensemble runs (10%-90% percentile and mean)
  - for all the observations in one area and the ensemble runs (10%-90% percentile and mean)

## All the imports

```python
import glob

#To deal with data
import xarray as xr

#For basic calculus
import numpy as np
import numpy.ma as ma

#For the plots
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import matplotlib.gridspec as gridspec

import pylab as plot
params = {'legend.fontsize': 20,
          'legend.handlelength': 2,
          'axes.titlesize' : 24,
          'axes.labelsize' : 20,
          'lines.linewidth' : 3,
          'lines.markersize' : 10,
          'xtick.labelsize' : 20,
          'ytick.labelsize' : 20}

plot.rcParams.update(params)
```

## Retrieve the data

```python
#the path where all data are stored
path_occiput_obs='/mnt/alberta/equipes/IGE/meom/MODEL_SET/ORCA025.L75/ORCA025.L75-OCCITENS/OBS'
#the specific path where data are extracted for year 2005
path_detar=path_occiput_obs+'/OBS-enact_y2005'
#the path to the file for one run (the 1st)
one_file=path_detar+'/ORCA025.L75-OCCITENS.001_y2005_enact_fdbk.nc'
```

```python
#opening one file
ds_one=xr.open_dataset(one_file,decode_times=False)
```

## Some functions

```python
def mask_one_profile(var):
    var_mask=ma.masked_where(var==9.99990000e+04,var)
    return var_mask
```

```python
def plots_temp_salt_model_obs_profiles(ds,n_obs):
    
    one_profile_lat=ds.LATITUDE[n_obs]
    one_profile_lon=ds.LONGITUDE[n_obs]
    one_profile_time=ds.JULD[n_obs]
    one_profile_temp_obs=ds.POTM_OBS[n_obs]
    one_profile_salt_obs=ds.PSAL_OBS[n_obs]
    one_profile_depth=ds.DEPTH[n_obs]
    one_profile_temp_mod=ds.POTM_Hx[n_obs]
    one_profile_salt_mod=ds.PSAL_Hx[n_obs]
    
    

    fig=plt.figure(figsize=(20,15))
    ax = plt.subplot(111,projection=ccrs.PlateCarree(central_longitude=0))
    ax.set_extent((one_profile_lon-20, one_profile_lon+20,
                   one_profile_lat-20, one_profile_lat+20))
    ax.coastlines(resolution='50m', color='black', linewidth=1)
    gl = ax.gridlines(draw_labels=True, linestyle=':', color='black',
                              alpha=0.5)
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    ax.tick_params('both',labelsize=22)

    plt.scatter(one_profile_lon,one_profile_lat,c='turquoise',linewidth=2, s=25,
                transform=ccrs.PlateCarree(),zorder=1)
    plt.title('Location of obs n '+str(n_obs))
    plt.savefig('plots/map_profile_n'+str(n_obs)+'.png')
    
    
    one_profile_temp_mod_mask=mask_one_profile(one_profile_temp_mod)
    one_profile_salt_mod_mask=mask_one_profile(one_profile_salt_mod)
    one_profile_temp_obs_mask=mask_one_profile(one_profile_temp_obs)
    one_profile_salt_obs_mask=mask_one_profile(one_profile_salt_obs)
    one_profile_depth_mask=mask_one_profile(one_profile_depth)
    
    gs = gridspec.GridSpec(1, 2, width_ratios=[1,1], height_ratios=[2])
    fig = plt.figure(figsize=(17, 15))  # (w,h)

    ax1 = plt.subplot(gs[0,0])

    ax1.plot(one_profile_temp_mod_mask,one_profile_depth_mask,'b.-', label='model')
    ax1.set_xlabel('Potential Temperature [$^\circ$C]')
    ax1.set_ylabel('Depth [m]')
    ax1.set_ylim(1000, 0)
    ax1.grid(True, which='both')
    ax1.xaxis.tick_top()
    ax1.xaxis.set_label_position('top') 
    ax1.plot(one_profile_temp_obs_mask,one_profile_depth_mask,'k.-', label='obs')
    ax1.legend()

    ax2 = plt.subplot(gs[0,1])

    ax2.plot(one_profile_salt_mod_mask,one_profile_depth_mask,'b.-', label='model')
    ax2.set_xlabel('Salinity PSU')
    ax2.set_ylabel('Depth [m]')
    ax2.set_ylim(1000, 0)
    ax2.grid(True, which='both')
    ax2.xaxis.tick_top()
    ax2.xaxis.set_label_position('top') 
    ax2.plot(one_profile_salt_obs_mask,one_profile_depth_mask,'k.-', label='obs')
    ax2.legend()
    
    plt.suptitle('Profiles for obs n '+str(n_obs), size=30)
    plt.savefig('plots/profiles_temp_salt_model_run1_obs_for_profile_n'+str(n_obs)+'.png')
```

## The plots

```python
plots_temp_salt_model_obs_profiles(ds_one,10)
```

## For all runs and one observation

### New function

```python
def plots_temp_salt_model_ensemble_obs_profiles(n_obs):
    
    #we open the first run
    file = path_detar+'/ORCA025.L75-OCCITENS.001_y2005_enact_fdbk.nc'
    ds=xr.open_dataset(file,decode_times=False)
    one_profile_lat=ds.LATITUDE[n_obs]
    one_profile_lon=ds.LONGITUDE[n_obs]
    one_profile_time=ds.JULD[n_obs]
    one_profile_temp_obs=ds.POTM_OBS[n_obs]
    one_profile_salt_obs=ds.PSAL_OBS[n_obs]
    one_profile_depth=ds.DEPTH[n_obs]
    one_profile_temp_mod=ds.POTM_Hx[n_obs]
    one_profile_salt_mod=ds.PSAL_Hx[n_obs]
    #we create an array for all runs to be stored
    profile_temp_mod_all_runs=np.zeros((50,len(one_profile_temp_mod)))
    profile_salt_mod_all_runs=np.zeros((50,len(one_profile_temp_mod)))
    #we fill them with the first value
    profile_temp_mod_all_runs[0]=one_profile_temp_mod
    profile_salt_mod_all_runs[0]=one_profile_salt_mod
    #we fill them with the rest of the runs 
    for n_run in np.arange(2,51):
        nn = "{:03d}".format(n_run)
        file = path_detar+'/ORCA025.L75-OCCITENS.'+str(nn)+'_y2005_enact_fdbk.nc'
        ds=xr.open_dataset(file,decode_times=False)
        one_profile_temp_mod=ds.POTM_Hx[n_obs]
        one_profile_salt_mod=ds.PSAL_Hx[n_obs]
        profile_temp_mod_all_runs[n_run-1]=one_profile_temp_mod
        profile_salt_mod_all_runs[n_run-1]=one_profile_salt_mod
        
    #we compute mean and percentiles
    mean_profile_temp_mod_all_runs=np.nanmean(profile_temp_mod_all_runs,axis=0)
    percent10_profile_temp_mod_all_runs=np.nanpercentile(profile_temp_mod_all_runs,10,axis=0)
    percent90_profile_temp_mod_all_runs=np.nanpercentile(profile_temp_mod_all_runs,90,axis=0)
    
    mean_profile_salt_mod_all_runs=np.nanmean(profile_salt_mod_all_runs,axis=0)
    percent10_profile_salt_mod_all_runs=np.nanpercentile(profile_salt_mod_all_runs,10,axis=0)
    percent90_profile_salt_mod_all_runs=np.nanpercentile(profile_salt_mod_all_runs,90,axis=0)
    
    #we plot the map of the location
    fig=plt.figure(figsize=(20,15))
    ax = plt.subplot(111,projection=ccrs.PlateCarree(central_longitude=0))
    ax.set_extent((one_profile_lon-20, one_profile_lon+20,
                   one_profile_lat-20, one_profile_lat+20))
    ax.coastlines(resolution='50m', color='black', linewidth=1)
    gl = ax.gridlines(draw_labels=True, linestyle=':', color='black',
                              alpha=0.5)
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    ax.tick_params('both',labelsize=22)

    plt.scatter(one_profile_lon,one_profile_lat,c='turquoise',linewidth=2, s=25,
                transform=ccrs.PlateCarree(),zorder=1)
    plt.title('Location of obs n '+str(n_obs))
    plt.savefig('plots/map_profile_n'+str(n_obs)+'.png')
    
    #we mask the profiles
    mean_profile_temp_mod_all_runs_mask=mask_one_profile(mean_profile_temp_mod_all_runs)
    percent10_profile_temp_mod_all_runs_mask=mask_one_profile(percent10_profile_temp_mod_all_runs)
    percent90_profile_temp_mod_all_runs_mask=mask_one_profile(percent90_profile_temp_mod_all_runs)
    mean_profile_salt_mod_all_runs_mask=mask_one_profile(mean_profile_salt_mod_all_runs)
    percent10_profile_salt_mod_all_runs_mask=mask_one_profile(percent10_profile_salt_mod_all_runs)
    percent90_profile_salt_mod_all_runs_mask=mask_one_profile(percent90_profile_salt_mod_all_runs)
    one_profile_temp_obs_mask=mask_one_profile(one_profile_temp_obs)
    one_profile_salt_obs_mask=mask_one_profile(one_profile_salt_obs)
    one_profile_depth_mask=mask_one_profile(one_profile_depth)
    
    #we plot the enveloppe of the runs
    gs = gridspec.GridSpec(1, 2, width_ratios=[1,1], height_ratios=[2])
    fig = plt.figure(figsize=(17, 15))  # (w,h)

    ax1 = plt.subplot(gs[0,0])

    ax1.plot(mean_profile_temp_mod_all_runs_mask,one_profile_depth_mask,'b.-', label='ensemble mean')
    ax1.plot(percent10_profile_temp_mod_all_runs_mask,one_profile_depth_mask,'b-', label='ensemble percent10')
    ax1.plot(percent90_profile_temp_mod_all_runs_mask,one_profile_depth_mask,'b-', label='ensemble percent90')
    ax1.fill_betweenx(one_profile_depth_mask, percent10_profile_temp_mod_all_runs_mask,
                      x2=percent90_profile_temp_mod_all_runs_mask, alpha=0.2, facecolor='b')
    ax1.set_xlabel('Potential Temperature [$^\circ$C]')
    ax1.set_ylabel('Depth [m]')
    ax1.set_ylim(2000, 0)
    ax1.grid(True, which='both')
    ax1.xaxis.tick_top()
    ax1.xaxis.set_label_position('top') 
    ax1.plot(one_profile_temp_obs_mask,one_profile_depth_mask,'k.-', label='obs')
    ax1.legend()

    ax2 = plt.subplot(gs[0,1])

    ax2.plot(mean_profile_salt_mod_all_runs_mask,one_profile_depth_mask,'b.-', label='ensemble mean')
    ax2.plot(percent10_profile_salt_mod_all_runs_mask,one_profile_depth_mask,'b-', label='ensemble percent10')
    ax2.plot(percent90_profile_salt_mod_all_runs_mask,one_profile_depth_mask,'b-', label='ensemble percent90')
    ax2.fill_betweenx(one_profile_depth_mask, percent10_profile_salt_mod_all_runs_mask,
                      x2=percent90_profile_salt_mod_all_runs_mask, alpha=0.2, facecolor='b')
    ax2.set_xlabel('Salinity PSU')
    ax2.set_ylabel('Depth [m]')
    ax2.set_ylim(2000, 0)
    ax2.grid(True, which='both')
    ax2.xaxis.tick_top()
    ax2.xaxis.set_label_position('top') 
    ax2.plot(one_profile_salt_obs_mask,one_profile_depth_mask,'k.-', label='obs')
    ax2.legend()
    
    plt.suptitle('Profiles for obs n '+str(n_obs), size=30)
    plt.savefig('plots/profiles_temp_salt_model_all_runs_obs_for_profile_n'+str(n_obs)+'.png')
```

```python
plots_temp_salt_model_ensemble_obs_profiles(10)
```

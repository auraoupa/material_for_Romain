```python
def mask_one_profile(var):
    var_mask=ma.masked_where(var==9.99990000e+04,var)
    return var_mask
```

```python
def plots_temp_salt_model_obs_profiles(n_obs):
    
    one_profile_lat=ds.LATITUDE[n_obs]
    one_profile_lon=ds.LONGITUDE[n_obs]
    one_profile_time=ds.JULD[n_obs]
    one_profile_temp_obs=ds.POTM_OBS[n_obs]
    one_profile_salt_obs=ds.PSAL_OBS[n_obs]
    one_profile_depth=ds.DEPTH[n_obs]
    one_profile_temp_mod=ds.POTM_Hx[n_obs]
    one_profile_salt_mod=ds.PSAL_Hx[n_obs]
    
    

    fig=plt.figure(figsize=(20,15))
    ax = plt.subplot(111,projection=ccrs.PlateCarree(central_longitude=0))
    ax.set_extent((one_profile_lon-30, one_profile_lon+30, one_profile_lat-30, 
                   one_profile_lat+30))
    ax.coastlines(resolution='50m', color='black', linewidth=1)
    gl = ax.gridlines(draw_labels=True, linestyle=':', color='black',
                              alpha=0.5)
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    ax.tick_params('both',labelsize=22)

    plt.scatter(one_profile_lon,one_profile_lat,c='turquoise',linewidth=2, s=25,
                transform=ccrs.PlateCarree(),zorder=1)
    plt.savefig('plots/map_profile_n'+str(n_obs)+'.png')
    
    
    one_profile_temp_mod_mask=mask_one_profile(one_profile_temp_mod)
    one_profile_salt_mod_mask=mask_one_profile(one_profile_salt_mod)
    one_profile_temp_obs_mask=mask_one_profile(one_profile_temp_obs)
    one_profile_salt_obs_mask=mask_one_profile(one_profile_salt_obs)
    one_profile_depth_mask=mask_one_profile(one_profile_depth)
    
    gs = gridspec.GridSpec(1, 2, width_ratios=[1,1], height_ratios=[2])
    fig = plt.figure(figsize=(20, 15))  # (w,h)

    ax1 = plt.subplot(gs[0,0])

    ax1.plot(one_profile_temp_mod_mask,one_profile_depth_mask,'b.-', label='model')
    ax1.set_xlabel('Potential Temperature [$^\circ$C]')
    ax1.set_ylabel('Depth [m]')
    ax1.set_ylim(1000, 0)
    ax1.grid(True, which='both')
    ax1.xaxis.tick_top()
    ax1.xaxis.set_label_position('top') 
    ax1.plot(one_profile_temp_obs_mask,one_profile_depth_mask,'k.-', label='obs')
    ax1.legend()

    ax2 = plt.subplot(gs[0,1])

    ax2.plot(one_profile_salt_mod_mask,one_profile_depth_mask,'b.-', label='model')
    ax2.set_xlabel('Salinity PSU')
    ax2.set_ylabel('Depth [m]')
    ax2.set_ylim(1000, 0)
    ax2.grid(True, which='both')
    ax2.xaxis.tick_top()
    ax2.xaxis.set_label_position('top') 
    ax2.plot(one_profile_salt_obs_mask,one_profile_depth_mask,'k.-', label='obs')
    ax2.legend()
    
    plt.savefig('plots/profiles_temp_salt_model_obs_for_profile_n'+str(n_obs)+'.png')
```
